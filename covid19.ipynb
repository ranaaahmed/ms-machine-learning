```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output  
import pandas as pd
import plotly.express as px
```

```python
app = dash.Dash(__name__)
```

```python
countries = data['Country/Region'].dropna().unique()
```

```python
app.layout = html.Div([
    html.H1("COVID-19 Dashboard", style={'text-align': 'center'}),
    
    
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in countries],
        value=countries[0], 
        placeholder="Select a Country"
    ),
    
    
    dcc.Graph(id='confirmed-cases-graph'),
    dcc.Graph(id='deaths-cases-graph'),
    dcc.Graph(id='recovered-cases-graph'),
]) 
```

```python
from dash.dependencies import Input, Output

@app.callback(
    [Output('confirmed-cases-graph', 'figure'),
     Output('deaths-cases-graph', 'figure'),
     Output('recovered-cases-graph', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_country):
 
    filtered_data = data[data['Country/Region'] == selected_country]
    
   
    confirmed_fig = px.line(filtered_data, x='Date', y='Confirmed', 
                            title=f'Confirmed Cases in {selected_country}')
    
    deaths_fig = px.line(filtered_data, x='Date', y='Deaths', 
                         title=f'Deaths in {selected_country}')
    
    recovered_fig = px.line(filtered_data, x='Date', y='Recovered', 
                            title=f'Recovered Cases in {selected_country}')
    
    return confirmed_fig, deaths_fig, recovered_fig
```

```python
app.run_server(mode='inline')
```
